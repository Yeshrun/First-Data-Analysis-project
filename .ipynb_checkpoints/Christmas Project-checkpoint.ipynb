```python
from matplotlib import pyplot as plt
```

```python
mean = [8.32,0.53,0.27,2.54,0.09,15.87,46.47,9.97e-01,3.31,0.66,10.42,5.64]
```

```python
properties = ["(fixed)","(vol)","(citric)","(res)","(chlor)","(free)","(total)","(den)","(pH)","(sul)","(alc)","(qua)"]
```

```python
# print(plt.style.available)
plt.style.use('classic')
plt.xlabel("Properties")
plt.ylabel("Mean")
plt.title("Mean of Properties")
plt.plot(properties,mean,color="#5a7d9a",linewidth= 2,marker ="o")
plt.grid()
plt.show()
```

where (fixed) refers to fixed acidity,(vol) refers to volatile acidity,(citric) refers to citric acidity,(res) residual sugar,
(chlor) refers to chlorides,(free) refers to free sulfur dioxides,(total) is total sulfur dioxide,(den) is density,(pH) is pH,(sul) is sulphates
,(alc) is alcohol,(qua) is quality

```python
red_wine["fixed acidity"].value_counts(bins= 10)
```

We can hence say that out of the 1599 bottles of red wine analysed, the probability of choosing a red wine bottle with 
fixed acidity from 6.86 to 7.98 is about 34%, which is the highest.


```python
red_wine["pH"].value_counts(bins=10)
```

We can conclude here that about 35 out of every 100 bottles of the 1599 wine bottles analysed will have a pH from 3.428 to 3.375 excluding 3.375;
about 23 out of every 100 bottles will have a pH within the range of (3.121-3.248]
