CORRELATION BETWEEN FIXED ACIDITY AND QUALITY

```python
y = red_wine.groupby('fixed acidity')[['quality']].mean()
y
```

```python
red_wine.groupby('fixed acidity')[['quality']].mean().sort_values(['quality'],ascending=False)
```

```python
from matplotlib import pyplot as plt
quality = [red_wine["quality"]]
fixed_acidity= [red_wine["fixed acidity"]]
plt.xlabel("Quality")
plt.ylabel("Fixed acidity")
plt.title("Correlation between Fixed acidity and Quality")
plt.scatter(quality,fixed_acidity,alpha = 0.75)
plt.figure(figsize=(10,8))
```

This correlation data frame in Pandas illustrates that the quality of the red wine is not dependent 
on how high its fixed acidity is.

The scatter plot further reaffirms this.This is because if the points align in a vertical line, then there is no correlation

EXPLORING FREQUENCIES OF A DEFINED RANGE IN A CHOSEN PROPERTY USING HISTOGRAMS AS WELL AS PROBABILITIES OF CHOOSING A BOTTLE FROM THAT RANGE

(1) CHOSEN PROPERTY: 'FIXED ACIDITY'

We have to write a block of code that groups the fixed acidities of the red wine into some range and displays the frequencies of fixed acidities within that certain range.
Conversely it can be used to determine the min and max value of the fixed acidity within the dataset,albeit,
a secondary use for the code block

```python
red_wine["fixed acidity"].value_counts(bins= 5)
```

Conclusion: We can hence say that out of the 1599 bottles of red wine analysed, the probability of choosing a red wine bottle at random with 
fixed acidity from 6.86 to 9.12 is about 57%, which is the highest. Then the probability of choosing randomly, a bottle with a fixed acidity between 9.12 and 11.38 is about 19%. The chances of choosing a bottle at random within the range of 4.588 to 6.86 is 17%. About 6% and 0.75% probability that a bottle chosen at random is within the range of 11.38 to 13.64 and 13.64 to 15.9 respectively 



```python
fixed_acidity = [red_wine["fixed acidity"]]
bins = (4.5,4.9,5.3,5.7,6.1,6.5,6.9,7.3,7.7,8.1,8.5,8.9,9.3,9.7,10.1,10.5,10.9,11.3,11.7,12.1,12.5,12.9,13.3,13.7,14.1,14.5,14.9)
plt.xlabel("fixed acidity range")
plt.ylabel("Frequencies")
plt.hist(fixed_acidity,bins = bins,edgecolor="black",alpha=0.75)
```

The histogram tells us that the highest number of red wine bottles can be found within a fixed acidity range of (9.3-9.7)
while the lowest occurring percentile is of the range(14.5-13.9)
